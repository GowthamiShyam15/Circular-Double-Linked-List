# Circular DLL

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None
        
class CDLL:
    def __init__(self):
        self.head = None
        self.tail = None
        
    #insertion at beginning    
    def insert_beg(self,data):
        nb = Node(data)
        if self.head == None:
            self.head = nb
            self.tail = nb
            self.head.prev = nb
            self.head.next = nb
        else:
            nb.next = self.head
            self.head.prev = nb
            nb.prev = self.tail
            self.tail.next = nb
            self.head = nb
            
    #insertion at ending    
    def insert_end(self,data):
        ne = Node(data)
        if self.head == None:
            self.head = ne
            self.tail = ne
            self.head.prev = ne
            self.head.next = ne
        else:
            self.tail.next = ne
            ne.prev = self.tail
            self.tail = ne
            self.tail.next = self.head
            self.head.prev = self.tail
            
    #insertion at specified position 
    def insert_pos(self,pos,data):
        np = Node(data)
        if self.head == None:
            self.head = np
            self.tail = np
            self.head.prev = np
            self.head.next = np
        else:
            temp = self.head
            for i in range(pos-1):
                temp = temp.next
            np.next = temp.next
            np.prev = temp
            temp.next = np
            temp.next.prev = np
            
    #Deletion at beginning
    def delete_beg(self):
        if self.head == None:
            print("Empty")
        else:
            temp = self.head
            self.head = temp.next
            temp.next = None
            temp.prev = None
            self.head.prev = self.tail
            self.tail.next = self.head
            
    #Deletion at ending
    def delete_end(self):
        if self.head == None:
            print("Empty")
        else:
            temp = self.head
            while temp.next != self.tail:
                temp = temp.next 
            temp.next = self.head
            self.head.prev = self.tail
            self.tail.next = None
            self.tail.prev = None
            self.tail = temp
            
    #Deletion at Specific Position
    def delete_pos(self,pos):
        if self.head == None:
            print("No Node to Delete")
        else:
            temp = self.head.next
            prev = self.head
            for i in range(pos-1):
                temp = temp.next
                prev = prev.next
            prev.next = self.tail
            self.tail.prev = prev
            temp.prev = None
            temp.next = None
            
            
        
    def display(self):
        temp = self.head
        if temp is None:
            print("CLL is Empty")
        else:
            temp = self.head
            print(temp.data,"-->",end=" ")
            while temp.next != self.head:
                temp = temp.next
                print(temp.data,"-->",end=" ")
            print(temp.next.data)
            
L = CDLL()

n1 = Node(10)
L.head = n1
L.tail = n1
n1.prev = n1
n1.next = n1

n2 = Node(20)
L.tail.next = n2
n2.prev = n1
n1.prev = n2
n2.next = n1
L.tail = n2

# n2.prev = L.head
# n2.next = L.head
# L.head.prev= n2
# L.head.next = n2
# L.tail = n2

# n2.prev = n1
# n1.next = n2
# n2.next = L.head
# L.head.prev = n2
# L.tail = n2

n3 = Node(30)
n3.prev = n2
L.tail.next = n3
n3.next = L.head
L.head.prev = n3
L.tail = n3

#Creation of Nodes
print("\nAfter Creation of Nodes:")
#displaying
L.display()

#insertion at beginning
print("\nInsertion at Beginning:")
L.insert_beg(5)
#displaying
L.display()

#insertion at ending
print("\nInsertion at Ending:")
L.insert_end(40)
#displaying
L.display()

#insertion at specified position
print("\nInsertion at specified position:")
L.insert_pos(2,100)
#displaying
L.display()

#Deletion at beginning
print("\nAfter Deletion at Beginning:")
L.delete_beg()
#displaying
L.display()

#Deletion at ending
print("\nAfter Deletion at Ending:")
L.delete_end()
#displaying
L.display()

#Deletion at Specific Position
print("\nAfter Deletion at Specific Position:")
L.delete_pos(2)
#displaying
L.display()
```
